```python
def mapGenerator(difficulty):
    global gameDimensions, startingX, startingY, levelDifficulty, startingMatrix, matrix, x, y
    if difficulty == 'easy':
        gameDimensions = [10,10]
        moves = math.floor(levelDifficulty/2)+4
        startingX = 0
        startingY = 0
        startingMatrix = numpy.zeros((gameDimensions),dtype=int)
        
        buildx = copy.deepcopy(startingX)
        buildy = copy.deepcopy(startingY)
        
        startingMatrix[buildy,buildx]=2
        
        
        WallPlacement(moves,buildx,buildy)
            
        
    elif difficulty == 'medium':
        gameDimensions = [15,15]
        moves = levelDifficulty+5
        startingX = 0
        startingY = 0
        startingMatrix = numpy.zeros((gameDimensions),dtype=int)
        
        buildx = copy.deepcopy(startingX)
        buildy = copy.deepcopy(startingY)
        
        startingMatrix[buildy,buildx]=2
                
        WallPlacement(moves,buildx,buildy)   
        
    elif difficulty == 'hard':
        gameDimensions = [25,25]
        moves = levelDifficulty+6
        startingX = 0
        startingY = 0
        startingMatrix = numpy.zeros((gameDimensions),dtype=int)
        
        buildx = copy.deepcopy(startingX)
        buildy = copy.deepcopy(startingY)
        
        startingMatrix[buildy,buildx]=2
                
        WallPlacement(moves,buildx,buildy)
        
    matrix = copy.deepcopy(startingMatrix)
    x = copy.deepcopy(startingX)
    y = copy.deepcopy(startingY)
    levelDifficulty = levelDifficulty+1
    #print(matrix)

def WallPlacement(moves,buildx,buildy):
    global gameDimensions, startingX, startingY, levelDifficulty, startingMatrix, matrix, x, y
    direction = random.randint(0,1)
    i = 0
    path = numpy.zeros((gameDimensions),dtype=int)
    iterations = 0
    while i < moves and iterations<1000:


        #print(buildy,buildx)        

        if direction == 0: #Up and down
            new = random.randint(0,gameDimensions[0]-1)
            goodFlag = True
            while goodFlag:
                
                if new>buildy:
                    yrange = range(buildy,new)
                elif new<=buildy:
                    yrange = range(new,buildy)
                    
                pathBlocked = False
                for y in yrange:
                    if startingMatrix[y,buildx] == 1:
                        pathBlocked = True
                
                
                
                if abs(new-buildy)<=1 or path[new,buildx] == 1 or pathBlocked:
                    new = random.randint(0,gameDimensions[0]-1)
                else:
                    goodFlag = False
                    
                iterations = iterations+1
                if iterations>1000:
                    break
            #print(abs(new-buildy))

            
            startingMatrix[new, buildx] = 1
            if new>buildy:
                for y in range(buildy,new):
                    path[y,buildx]=1
                buildy = new-1
            elif new<buildy:
                for y in range(new,buildy):
                    path[y,buildx]=1
                buildy = new+1
                    
            direction = 1

            if i == moves-1:
                 startingMatrix[new, buildx] = 5

        elif direction == 1: #Right and left
            new = random.randint(0,gameDimensions[1]-1)
            
            goodFlag = True
            while goodFlag:
                
                if new>buildx:
                    xrange = range(buildx,new)
                elif new<=buildy:
                    xrange = range(new,buildx)
                    
                pathBlocked = False
                for x in xrange:
                    if startingMatrix[buildy,x] == 1:
                        pathBlocked = True
                        
                
                                        
                if abs(new-buildx)<=1 or path[buildy,new] == 1 or pathBlocked:
                    new = random.randint(0,gameDimensions[1]-1)
                else:
                    goodFlag = False
                    
                iterations = iterations+1
                
                if iterations>1000:
                    break 
            #print(abs(new-buildx))
            
            startingMatrix[buildy, new] = 1
            if new>buildx:
                for x in range(buildx,new):
                    path[buildy,x]=1
                buildx = new-1
            elif new<buildx:
                for x in range(new,buildx):
                    path[buildy,x]=1
                buildx = new+1
                    
            direction = 0
            
            if i == moves-1:
                 startingMatrix[buildy, new] = 5

        i = i+1
```
